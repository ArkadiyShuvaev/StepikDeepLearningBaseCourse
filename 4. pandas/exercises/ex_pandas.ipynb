```python
total_customers_count = len(data)
ageCategories = data.groupby(data["Age"])["Age"].unique()

filtered_customers = data[((data.Gender == "M") & (data.Age == "26-35")) | ((data.Gender == "F") & ((data.Age == "36-45") | (data.Age == "46-50") | (data.Age == "51-55") | (data.Age == "55+")))]
filtered_customers2 = data[((data.Gender == "M") & (data.Age == "26-35")) | ((data.Gender == "F") & (data.Age > "36"))]

filtered_customers_count = len(filtered_customers)
filtered_customers2_count = len(filtered_customers2)

proportion = filtered_customers_count / total_customers_count
print(proportion)

proportion2 = filtered_customers2_count / total_customers_count
print(proportion2)
```
